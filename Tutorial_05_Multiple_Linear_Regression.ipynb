```python
#4.
student.info()
```

```python
# Find Missing Values

student.isnull().sum()
```

```python
# Missing values Zero

student.fillna(student.mean(),inplace=True)
student.isnull().sum()
```

```python
# using Multiple linear regration

X = student.iloc[: , :-1]
X
```

```python
Y = student.iloc[: , -1]
Y
```

```python
#useing multiple regression

from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X , Y)
print("Intercept : - " , model.intercept_)

print("Coefficient : - " , model.coef_)
```
