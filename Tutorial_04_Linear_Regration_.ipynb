```python
#1.
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cm=pd.read_csv('/content/drive/MyDrive/SEM 06/computers.csv')
cm
```

```python
# Find Out Row And Column
cm.info()
```

```python
#Mean Values Form minutes
mean_value =cm['Minutes'].mean()
print(mean_value)
```

```python
plt.scatter(cm['Units'], cm['Minutes'],color='blue', marker='o')

plt.axhline(y=mean_value,c="r")

plt.xlabel('Units')
plt.ylabel('Minutes')

plt.show()
```

```python
#creation of 3 models
model_0 =cm['Minutes'].mean()
model_1=10 + 12*cm['Units']
model_2=6 + 18*cm['Units']

model_2
```

```python
# Data Table For Model 0 , 1 , 2
cm['Model_0']=model_0
cm['Model_1']=model_1
cm['Model_2']=model_2
cm
```

```python
fig,ax=plt.subplots()

ax.scatter(x='Units' , y= "Minutes" ,data=cm, label="Actual Repair Time")

ax.plot(cm['Units'], cm['Model_0'], label="Model 0 (Mean)", color="red" )
ax.plot(cm['Units'], cm['Model_1'], label="Model 1 (12 + 10 * Units)", color="green" )
ax.plot(cm['Units'], cm['Model_2'], label="Model 2 (6 + 18 * Units)", color="blue")

ax.set_xlabel('Units')
ax.set_ylabel('Minutes')
ax.set_title('Speculated Models')
ax.legend()
```

```python
plt.figure(figsize=(10, 6))

# Original data points
plt.scatter(cm['Units'], cm['Minutes'], color="blue", label="Original Data", s=50, zorder=3)

plt.plot(cm['Units'], cm['Model_0'], label="Model 0 (Mean)", color="green", linestyle="--")

plt.plot(cm['Units'], cm['Model_1'], label="Model 1 (12 + 10 * Units)", color="orange")

plt.plot(cm['Units'], cm['Model_2'], label="Model 2 (6 + 18 * Units)", color="red")
# Plot each model
plt
```

```python
#Build Model Using Skytlearn

x=cm[['Units']]

y=cm['Minutes']

from sklearn.linear_model import LinearRegression

model=LinearRegression()

model.fit(x,y)

print("Intercept : - " , model.intercept_)

print("Coefficient : - " , model.coef_)
```

```python
model_0 =cm['Minutes'].mean()
model_1=10 + 12*cm['Units']
model_2=6 + 18*cm['Units']
model_3=model.intercept_ + model.coef_*cm['Units']
```

```python
# Data Table For Model 0 , 1 , 2
cm['Model_0']=model_0
cm['Model_1']=model_1
cm['Model_2']=model_2
cm['Model_3']=model_3
cm
```

```python
plt.figure(figsize=(10, 6))

# Original data points
plt.scatter(cm['Units'], cm['Minutes'], color="blue", label="Original Data", s=50, zorder=3)

plt.plot(cm['Units'], cm['Model_0'], label="Model 0 (Mean)", color="green", linestyle="--")

plt.plot(cm['Units'], cm['Model_1'], label="Model 1 (12 + 10 * Units)", color="orange")

plt.plot(cm['Units'], cm['Model_2'], label="Model 2 (6 + 18 * Units)", color="red")

plt.plot(cm['Units'], model_3, label="Best Fit Model", color="purple", linestyle="-")
# Plot each model
plt.xlabel('Units')
plt.ylabel('Minutes')
plt.title('Speculated Models')
plt.legend()
plt.show()
```

```python
#Using mean square error find accuracy R2 Values

rsq = model.score(cm[['Units']],y)
rsq
```
