```python
pip install ucimlrepo
```

```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
heart_disease = fetch_ucirepo(id=45)

# data (as pandas dataframes)
X = heart_disease.data.features
y = heart_disease.data.targets

# metadata
print(heart_disease.metadata)

# variable information
print(heart_disease.variables)
```

```python
# Rows with missing values in X
rows_with_missing_X = X[X.isnull().any(axis=1)]
print("Rows with missing values in features (X):")
print(rows_with_missing_X)

# Rows with missing values in y
rows_with_missing_y = y[y.isnull()]
print("Rows with missing values in targets (y):")
print(rows_with_missing_y)
```
