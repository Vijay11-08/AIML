# Data Preprocessing
- Machine Learning Algoritham can work with numerical data only
- During feture enginerning the task of converting categorial fetures into numerical data is a **Encodeing**  
- Ther are differnt methods convert categorial data into numerical data is **Encoding**.
    1. Leble Encoding
       - Assign a Unique Integer to each category.
       - [ ' cat ' , ' dog ' ] --> [ 0 , 1 ]
    2. One-Hot Encoding
       - [ ' Red ' , ' Green ' , ' Blue ' ] --> [ 1 ,0 , 0 ] , [ 0 , 1 , 0 ] and [ 0 , 0 , 1 ]
    3. Ordinal Encoding
       - Assign a integer to categories based on a defiened order.
       - [ a , b , c , d , e ] --> [ 1 , 2 , 3 , 4 , 5 ]
